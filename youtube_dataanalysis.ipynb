```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/AAA.csv", encoding='ISO-8859-1')
```

```python
df
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.head()
```

```python
df['lowest_monthly_earnings'] = df['lowest_monthly_earnings'].replace('[\$,]', '', regex=True).astype(float)
df['highest_monthly_earnings'] = df['highest_monthly_earnings'].replace('[\$,]', '', regex=True).astype(float)
df['lowest_yearly_earnings'] = df['lowest_yearly_earnings'].replace('[\$,]', '', regex=True).astype(float)
df['highest_yearly_earnings'] = df['highest_yearly_earnings'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df
```

```python
from pandas.tseries.offsets import DateOffset
df['created_date'] = pd.to_datetime(df['created_date'])
now = pd.to_datetime('now')
df['channel_age_years'] = now.year - df['created_date'].dt.year
df['channel_age_months'] = now.month - df['created_date'].dt.month
df['channel_age_months'] = df['channel_age_years'] * 12 + df['channel_age_months']
print(df[['created_date', 'channel_age_months']].head())
```

```python
df['subscribers_per_video'] = df['subscribers'] / df['uploads']
df['views_per_video'] = df['video views'] / df['uploads']
```

```python
q1 = df['subscribers'].quantile(0.25)
q3 = df['subscribers'].quantile(0.75)
iqr = q3 - q1
df = df[~((df['subscribers'] < (q1 - 1.5 * iqr)) | (df['subscribers'] > (q3 + 1.5 * iqr)))]
```

```python
top_10_channels = df[['Youtuber', 'subscribers']].sort_values(by='subscribers', ascending=False).head(10)
print("Top 10 YouTube Channels by Subscribers:")
print(top_10_channels)
```

```python
avg_subscribers_per_category = df.groupby('category')['subscribers'].mean().sort_values(ascending=False)
print("Category with the Highest Average Number of Subscribers:")
print(avg_subscribers_per_category)
```

```python
avg_uploads_per_category = df.groupby('category')['uploads'].mean().sort_values(ascending=False)
print("Average Number of Videos Uploaded per Category:")
print(avg_uploads_per_category)
```

```python
top_5_countries = df['Country'].value_counts().head(5)
print("Top 5 Countries by Number of YouTube Channels:")
print(top_5_countries)
```

```python
channel_type_distribution = pd.crosstab(df['channel_type'], df['category'])
print("Distribution of Channel Types Across Categories:")
print(channel_type_distribution)
```

```python
correlation_subscribers_views = df['subscribers'].corr(df['video views'])
print(f"Correlation between the number of subscribers and total video views: {correlation_subscribers_views}")
```

```python
earnings_by_category = df.groupby('category')[['lowest_monthly_earnings', 'highest_monthly_earnings']].mean()
print("Monthly earnings variation by category:")
print(earnings_by_category)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='created_date', y='subscribers_for_last_30_days', data=df)
plt.title('Trend in Subscribers Gained in the Last 30 Days')
plt.xlabel('Channel Creation Date')
plt.ylabel('Subscribers Gained in Last 30 Days')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='lowest_yearly_earnings', data=df)
plt.title('Outliers in Lowest Yearly Earnings')
plt.xlabel('Lowest Yearly Earnings')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='highest_yearly_earnings', data=df)
plt.title('Outliers in Highest Yearly Earnings')
plt.xlabel('Highest Yearly Earnings')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['created_date'].dt.year.value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Channel Creation Dates')
plt.xlabel('Year of Creation')
plt.ylabel('Number of Channels')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Gross tertiary education enrollment (%)', y='Country', data=df)
plt.title('Tertiary Education Enrollment vs. Number of Channels')
plt.xlabel('Gross Tertiary Education Enrollment (%)')
plt.ylabel('Country')
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(10).index
unemployment_top_countries = df[df['Country'].isin(top_countries)][['Country', 'Unemployment rate']]
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Unemployment rate', data=unemployment_top_countries)
plt.title('Unemployment Rate in Top 10 Countries by Number of YouTube Channels')
plt.xlabel('Country')
plt.ylabel('Unemployment Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_urban_population = df['Urban_population'].mean()
print(f"Average Urban Population Percentage in Countries with YouTube Channels: {avg_urban_population:.2f}%")
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=df)
plt.title('Distribution of YouTube Channels Based on Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
correlation_subscribers_population = df['subscribers'].corr(df['Population'])
print(f"Correlation between the number of subscribers and the population of the country: {correlation_subscribers_population}")
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='channel_type', y='video_views_for_the_last_30_days', data=df)
plt.title('Distribution of Video Views for the Last 30 Days Across Channel Types')
plt.xlabel('Channel Type')
plt.ylabel('Video Views in Last 30 Days')
plt.xticks(rotation=45)
plt.show()
```

```python
df = df.copy()
df['created_month'] = df['created_date'].dt.month
avg_uploads_per_month = df.groupby('created_month')['uploads'].mean()
plt.figure(figsize=(12, 6))
avg_uploads_per_month.plot(kind='line')
plt.title('Seasonal Trends in Number of Videos Uploaded by YouTube Channels')
plt.xlabel('Month')
plt.ylabel('Average Number of Uploads')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
